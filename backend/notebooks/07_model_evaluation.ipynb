```python
import sys
import os

project_root = os.path.abspath("..")
if project_root not in sys.path:
    sys.path.append(project_root)

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from src.preprocessing import balance_training_data
from src.models import (
    logistic_model,
    random_forest_model,
    xgboost_model,
    stacking_model
)
from src.evaluation import compare_models

# Load Data
X_train = pd.read_csv("../data/X_train_features.csv")
X_test = pd.read_csv("../data/X_test_features.csv")

y_train = pd.read_csv("../data/y_train.csv").squeeze()
y_test = pd.read_csv("../data/y_test.csv").squeeze()

X_train_bal, y_train_bal = balance_training_data(X_train, y_train)

models = {
    "Logistic": logistic_model().fit(X_train_bal, y_train_bal),
    "Random Forest": random_forest_model().fit(X_train_bal, y_train_bal),
    "XGBoost": xgboost_model().fit(X_train_bal, y_train_bal),
    "Stacking": stacking_model().fit(X_train_bal, y_train_bal)
}

results = compare_models(models, X_test, y_test)
print(results)

# Recall Comparison
plt.figure()
sns.barplot(x="Model", y="Recall", data=results)
plt.title("Recall Comparison")
plt.xticks(rotation=45)
plt.show()

# Precision Comparison
plt.figure()
sns.barplot(x="Model", y="Precision", data=results)
plt.title("Precision Comparison")
plt.xticks(rotation=45)
plt.show()

# F1 Comparison
plt.figure()
sns.barplot(x="Model", y="F1-Score", data=results)
plt.title("F1 Score Comparison")
plt.xticks(rotation=45)
plt.show()

# ROC-AUC Comparison
plt.figure()
sns.barplot(x="Model", y="ROC-AUC", data=results)
plt.title("ROC-AUC Comparison")
plt.xticks(rotation=45)
plt.show()
```
