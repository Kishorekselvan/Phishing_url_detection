```python
import sys
import os
import joblib
# Add project root to path
project_root = os.path.abspath("..")
if project_root not in sys.path:
    sys.path.append(project_root)

import pandas as pd
from src.feature_engineering import transform_urls

# ==============================
# 1️⃣ Load Train & Test URLs
# ==============================
X_train = pd.read_csv("../data/X_train.csv")['url']
X_test = pd.read_csv("../data/X_test.csv")['url']

# Optional sanity check
print("Sample URLs:")
print(X_train.head())

# Remove invalid rows (extra safety)
X_train = X_train.dropna()
X_test = X_test.dropna()

X_train = X_train[X_train.str.len() > 3]
X_test = X_test[X_test.str.len() > 3]

# ==============================
# 2️⃣ Feature Extraction
# ==============================
print("\nExtracting features...")

X_train_features = transform_urls(X_train)
X_test_features = transform_urls(X_test)

# ==============================
# 3️⃣ Verify Features
# ==============================
print("\nFeature shape:", X_train_features.shape)
print("\nFeature columns:")
print(X_train_features.columns)

print("\nSample Features:")
print(X_train_features.head())

# ==============================
# 4️⃣ Save Features
# ==============================
X_train_features.to_csv("../data/X_train_features.csv", index=False)
X_test_features.to_csv("../data/X_test_features.csv", index=False)

print("\nFeature Extraction Completed Successfully ✅")
import joblib

os.makedirs("../saved_models", exist_ok=True)

feature_columns = list(X_train_features.columns)

joblib.dump(feature_columns, "../saved_models/feature_columns.pkl")

print("\nFeature columns saved successfully ✅")
```
